## 1. Compare circular velocity curves with figures from Batsleer & Dejonghe 1994

Setup parameters of the first row of figures in fig. 1, 2 and 3 

```python
#surface ratios of disk and halo:
ac_Ds = [50.  ,50.  ,50.  ,50. ,50. ,50.  ,40. ,40. ,40. ]
ac_Hs = [1.005,1.005,1.005,1.01,1.01,1.01 ,1.02,1.02,1.02 ]
#disk contribution to total mass:
ks    = [0.05 ,0.06 ,0.07 ,0.07,0.1 ,0.125,0.1 ,0.12,0.125]
#focal distance:
Delta = 1.
```

Setup 2-component KKSPot and plot rotation cuve for each parameter set and compare with formula by Batsleer & Dejonghe

```python
%matplotlib inline

#_____import packages_____
from galpy.potential import KuzminKutuzovStaeckelPotential
from galpy.potential import evaluatePotentials, evaluateRforces
import matplotlib.pyplot as plt
import numpy

fig = plt.figure(figsize=(8,6))

for ii in range(9):
    ac_D = ac_Ds[ii]
    ac_H = ac_Hs[ii]
    k    = ks[ii]
    
    #_____setup potential_____
    #first try, not normalized:
    amp_D = k
    V_D = KuzminKutuzovStaeckelPotential(amp=amp_D,ac=ac_D,Delta=Delta,normalize=False)
    amp_H = 1.-k
    V_H = KuzminKutuzovStaeckelPotential(amp=amp_H,ac=ac_H,Delta=Delta,normalize=False)
    pot = [V_D,V_H]
    #normalization according to Batsleer & Dejonghe 1994:
    V00 = evaluatePotentials(0.,0.,pot)
    #second try, normalized:
    amp_D = k / (-V00)
    V_D = KuzminKutuzovStaeckelPotential(amp=amp_D,ac=ac_D,Delta=Delta,normalize=False)
    amp_H = (1.-k) / (-V00)
    V_H = KuzminKutuzovStaeckelPotential(amp=amp_H,ac=ac_H,Delta=Delta,normalize=False)
    pot = [V_D,V_H]
    
    #_____calculate rotation curve_____
    Rs = numpy.linspace(0.,20.,100)
    z = 0.
    vcirc   = numpy.sqrt(-Rs * evaluateRforces(Rs,z,pot))
        
    #_____vcirc by Batsleer & Dejonghe eq. (10) (with proper Jacobian)_____
    def vc2w(R):
        g_D = Delta**2 / (1.-ac_D**2)
        a_D = g_D - Delta**2
        g_H = Delta**2 / (1.-ac_H**2)
        a_H = g_H - Delta**2
        l = R**2 - a_D
        q = a_H - a_D
        termD = numpy.sqrt(l  )*(numpy.sqrt(l  ) + numpy.sqrt(-g_D  ))**2
        termH = numpy.sqrt(l-q)*(numpy.sqrt(l-q) + numpy.sqrt(-g_D-q))**2
        return R**2 * (k / termD + (1.-k) / termH)
    vcircBD = numpy.sqrt(vc2w(Rs)/(-V00))
    
    #_____plot_____
    ax = plt.subplot(3,3,ii+1)
    ax.plot(Rs,vcircBD,color='red' ,linestyle='solid' ,label='BD94')
    ax.plot(Rs,vcirc,  color='blue',linestyle='dashed',label='KKSPot')
    ax.set_ylim([0.,0.52])
    
    #_____format_____
    if ii == 8: ax.legend(loc=4)
    if ii > 5: plt.xlabel("R")
    if numpy.any(numpy.array([0,3,6]) == ii): plt.ylabel("$v_{circ}$ in Fig."+str(ii/3+1)+", first row") 
plt.tight_layout()
```

Result: Producing circular velocity curves works!

## 3. Comparing python & C implementation by calculating an orbit

```python
#_____import packages_____
from galpy.potential import KuzminKutuzovStaeckelPotential
from galpy.potential import evaluatePotentials, evaluateRforces
import matplotlib.pyplot as plt
import numpy
import sys
from galpy.orbit import Orbit

#_____initialize some KKSPot_____
Delta = 1.0
pot = KuzminKutuzovStaeckelPotential(ac=20.,Delta=Delta,normalize=True)

#_____initialize an orbit (twice)_____
vxvv = [1.,0.1,1.1,0.,0.1]
o_P= Orbit(vxvv=vxvv)
o_C= Orbit(vxvv=vxvv)

#_____integrate the orbit with python and C_____
ts= numpy.linspace(0,100,10000)
o_P.integrate(ts,pot,method='leapfrog')  #python
o_C.integrate(ts,pot,method='leapfrog_c')#C

#_____figure_____
o_P.plot(overplot=False,color='red',linestyle='solid',label='python')
o_C.plot(overplot=True,color='blue',linestyle='dashed',label='c')
plt.legend()

o_P.plot(overplot=False,color='red',linestyle='solid',label='python',d1='vR',d2='vz')
o_C.plot(overplot=True,color='blue',linestyle='dashed',label='c',d1='vR',d2='vz')
```

Result: Works!

## 6. Test 3-Component Stäckel potential from Famaey & Dejonghe (2003)

As an additional test I also tried to reproduce plots from Famaey & Dejonghe (2003), which use the same potential, but with three instead of only 2 components.

```python
#_____setup parameters for potentials in Famaey & Dejonghe (2003), tab. 1_____
ac_thin_s  = [75.,200.,200.,75.,200.]
ac_thick_s = [1.5,1.8,1.3,1.8,1.3]
ac_halo_s  = [1.02,1.02,1.01,1.01,1.005]
k_thin_s   = [0.13,0.10,0.11,0.11,0.07]
k_thick_s  = [0.01,0.01,0.04,0.015,0.01]
scale_rs         = [0.93,0.88,0.95,0.98,1.01]
Mmin       = [2.37,2.37,3.19,3.56,6.13]
Mmax       = [2.41,2.41,3.22,3.58,6.20]
Mtot       = 0.5*(numpy.array(Mmin)+numpy.array(Mmax))
```

Reproduce fig. 5, circular velocity curve:

```python
#_____import packages_____
from galpy.potential import evaluateRforces



fig = plt.figure(figsize=(6,6))

for ii in [0,1,2,3,4]:
    ac_thin  = ac_thin_s[ii]
    ac_thick = ac_thick_s[ii]
    ac_halo  = ac_halo_s[ii]
    k_thin   = k_thin_s[ii]
    k_thick  = k_thick_s[ii]
    Delta    = scale_rs[ii]
    GM       = 430000.*Mtot[ii] #units: [(km/s)^2*kpc]
    
    #_____setup potential_____
    amp_thin = GM * k_thin
    V_thin = KuzminKutuzovStaeckelPotential(amp=amp_thin,ac=ac_thin,Delta=Delta,normalize=False)
    amp_thick = GM * k_thick
    V_thick = KuzminKutuzovStaeckelPotential(amp=amp_thick,ac=ac_thick,Delta=Delta,normalize=False)
    amp_halo = GM * (1.-k_thin-k_thick)
    V_halo = KuzminKutuzovStaeckelPotential(amp=amp_halo,ac=ac_halo,Delta=Delta,normalize=False)
    pot = [V_thin,V_thick,V_halo]
        
    #_____calculate rotation curve_____
    Rs = numpy.linspace(0.,20.,100)
    z = 0.
    vcirc   = numpy.sqrt(-Rs * evaluateRforces(Rs,z,pot)) #units: 
    
    #_____plot rotation curve_____
    plt.plot(Rs,vcirc)
    plt.axis([-1.,21.,-10.,280.])
    plt.xlabel("galactocentric radius [kpc]")
    plt.ylabel("circular velocity [km/s]")
```

The plot above looks like fig. 5 in Famaey & Dejonghe (2003). I therefore got the units and normalisation right. Let's try to reproduce fig. 4. Fig. 4 shows the density profile for two of the above potentials. I believe that the density is given in units of solar mass per pc^3.
